```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("/content/drive/MyDrive/creditcard.csv")
```

```python
# Data is highly unbalance 0-> Normal Transaction and 1-> Fraudulent Transaction
```

```python
# Separating the data for analysis
```

```python
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

```python
legit['Amount'].describe()
```

```python
fraud['Amount'].describe()
```

```python
# Compare the values of both transaction
df.groupby('Class').mean()
```

```python
legit_sample=legit.sample(n=492)
```

```python
new_df=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
X=new_df.drop(columns='Class',axis=1)
Y=new_df['Class']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X_train.shape,X_test.shape,X.shape)
```

```python
print(Y_train.shape,Y_test.shape,Y.shape)
```

```python
# MOdel Training
# Logistic Regression
```

```python
model=LogisticRegression()
```

```python
#training the data
model.fit(X_train,Y_train)
```

```python
# MOdel Evaluation
# Accuracy Score
x_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(x_train_prediction,Y_train)
```

```python
print('Accuracy on the training data: ',training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy score of testing data: ",testing_data_accuracy)
```
